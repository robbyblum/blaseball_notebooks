## Decrees

Decrees are chosen by majority votes across all teams. `2` decrees will pass this season.

### -Sun

**Set Sun. Non-losing with Negative Runs earns you 2 Wins.**

See below for the number of additional Wins this would have added. It is unknown how this interacts with the current
Turntables/SunSun mess.

```python
extra_wins = 0
s21_games = Game.load_by_season(season=21, team_id=pies.id)
for g in s21_games.values():
    if g.winning_team.id == pies.id and g.winning_score < 0:
        extra_wins += 1
print(f"Pies would have an extra {extra_wins} Wins")
```

### Sun 0

**Set Sun. Non-losing with Zero Runs earns you 2 Wins.**

See below for the number of additional Wins this would have added. It is unknown how this interacts with the current
Turntables/SunSun mess.

```python
extra_wins = 0
s21_games = Game.load_by_season(season=21, team_id=pies.id)
for g in s21_games.values():
    if g.winning_team.id == pies.id and g.winning_score == 0:
        extra_wins += 1
print(f"Pies would have an extra {extra_wins} Wins")
```

### Equal Sun

**Set Sun. A home run during a game's final out ties the game.**

This will, presumably, force a game to go into extra innings if the final play is a home run. This will increase the
number of Sun 30 Wins awarded.

### MAXIMUM SUN

**Set Sun. Scoring plays in Maximum Blaseball are worth 3x the Runs.**

Maximum Blaseball occurs when all the bases are loaded and there are 2 outs, 2 strikes, and 3 balls in the count. This
is a fairly rare occurrence. Note that some players also have Minimalist or Maximalist, making them play better or worse
in this situation

### Sum Sun

**Weather. Each successive scoring play by a Team within a half inning is worth an additional Run.**

This weather will reward chaining together many running scores (Whole Team-ing).

### Sun 9

**Set Sun. Non-losses on the Week End (every 9th Day) are worth 2 Wins.**

See below for the number of additional Wins this would have added. It is unknown how this interacts with the current
Turntables/SunSun mess.

```python
extra_wins = 0
for g in s21_games.values():
    if (g.day - 1) % 9 == 8 and g.day <= 99:
        if g.winning_team.id == pies.id:
            extra_wins += 1
print(f"Pies would have an extra {extra_wins} Wins")
```

## Wills

Will are selected for each team by raffle. Each team will draw `2` random wills. The bottom 4 teams in the league by
standings will draw `3`. The chosen wills are random, but wills with more votes are more likely to be selected.


### Equivalent Exchange
**Swap a Player from a Team in the League with an Equivalent (within 2 Stars) Player on your Team.**

See below for the 20 best swap candidates for a particular player, sorted by highest stats in the applicable position.

```python
league = League.load()
all_teams = list(league.teams.values())
all_active = []
for team in all_teams:
    all_active.extend(team.lineup + team.rotation)
all_active_league = [x for x in all_active if x.league_team.id != pies.id]
all_total_stars = get_total_stars(all_active_league)
```

```python
for player in pies.lineup:
    total_stars = get_total_stars(player)[player.id]
    valid_total_stars = {k: v for k, v in all_total_stars.items() if total_stars-2 < v < total_stars+2}
    player_matches = [x for x in all_active_league if x.id in valid_total_stars.keys()]
    player_matches.sort(key=lambda x: x.batting_rating, reverse=True)
    display(Markdown(f"####{player.name}"))
    display(get_stars([player] + player_matches[0:20], include_team=True))
    display(set_heatmap(get_batting_stlats([player] + player_matches[0:20]), maxVal=1.5))
```

```python
for player in pies.rotation:
    total_stars = get_total_stars(player)[player.id]
    valid_total_stars = {k: v for k, v in all_total_stars.items() if total_stars-2 < v < total_stars+2}
    player_matches = [x for x in all_active_league if x.id in valid_total_stars.keys()]
    player_matches.sort(key=lambda x: x.pitching_rating, reverse=True)
    display(Markdown(f"####{player.name}"))
    display(get_stars([player] + player_matches[0:20], include_team=True))
    display(set_heatmap(get_pitching_stlats([player] + player_matches[0:20]), maxVal=1.5))
```

### Gachapon

**Recruit one random Player from each other Team's Shadows to your Team's Shadows.**

See below for average stats for each team's shadows.

```python
averages = []
for team in all_teams:
    fake_player = {"hittingRating": mean([x.hitting_rating for x in team.shadows]),
                   "pitchingRating": mean([x.pitching_rating for x in team.shadows]),
                   "baserunningRating": mean([x.baserunning_rating for x in team.shadows]),
                   "defenseRating": mean([x.defense_rating for x in team.shadows]),
                   "name": team.nickname,
                   "id": team.id}
    averages.append(Player(fake_player))
get_stars(averages)
```
